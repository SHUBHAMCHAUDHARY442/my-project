#### Chart - 11

````python


import matplotlib.pyplot as plt
import seaborn as sns

# Calculate average ADR for City Hotel and Resort Hotel
avg_adr_city = df[df['hotel'] == 'City Hotel']['adr'].mean()
avg_adr_resort = df[df['hotel'] == 'Resort Hotel']['adr'].mean()

# Plotting the bar chart with average ADR values on top
plt.figure(figsize=(10, 6))
ax = sns.barplot(x=['City Hotel', 'Resort Hotel'], y=[avg_adr_city, avg_adr_resort], palette='pastel')

# Adding average ADR values on top of each bar
for p, value in zip(ax.patches, [avg_adr_city, avg_adr_resort]):
    ax.annotate(f'{value:.2f}', (p.get_x() + p.get_width() / 2., p.get_height()), ha='center', va='center', xytext=(0, 10), textcoords='offset points')

plt.title('Average ADR Comparison - City Hotel vs. Resort Hotel')
plt.xlabel('Hotel Type')
plt.ylabel('Average ADR')
plt.show()

````

##### 1. Why did you pick the specific chart?

In the context of visualizing the distribution of the 'ADR' (Average Daily Rate) across different hotel types ('City Hotel' and 'Resort Hotel')
